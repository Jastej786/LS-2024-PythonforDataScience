```python
''' d)Write a Python function numpy_spiral_order(matrix) that takes a NumPy matrix and returns
a list containing the elements visited in a spiral order.'''

def numpy_spiral_order(matrix):
  arr =np.empty(0,dtype = int)
  no_of_rows, no_of_columns = np.shape(matrix)
  for i in range(no_of_rows):
    if i % 2 == 0:
      x = matrix[i][::1]
    else:
      x = matrix[i][::-1]
    i = i+1
    arr = np.hstack((arr, x))
  return arr
```

```python
s = df['Genre'].str.count('Shooter')
s.sum()
```

```python
genre_count = np.empty(0,dtype = int)
for i in genre:
  genre_count = np.append(genre_count, df['Genre'].str.count(i).sum())
genre_count
```

```python
fig, ax = plt.subplots(figsize = (14,4))
ax.bar(genre, genre_count, color = 'purple')
plt.show()
```
